```python
material_mapping = {
    'Cotton': ['cotton', 'cotton_100', 'cotton_blend', 'cotton_mostly'],
    'Wool': ['wool', 'wool_blend', 'wool_mostly', 'virgin_wool','tweed'],
    'leather': ['leather', 'lather', 'down'],
    'faux_fur':['faux_fur'],
    'Polyester': ['polyester', 'polyester_100', 'polyester_mostly'],
    'Nylon': ['nylon', 'nylon_100'],
    'Polyamide': ['polyamide', 'polyamide_100', 'polamide'],
    'Other material': ['elastane','jacquard','lycra', 'ripstop','Compound','linen', 'linen_blend','viscose', 'viscose_100','satin'],
    'corduroy': ['corduroy'],
    'fleece':['fleece'],
    'rib_knit':['rib_knit'],
    'denim':['denim']
}

# Create reverse mapping (from material to category)
reverse_mapping = {}
for category, materials in material_mapping.items():
    for material in materials:
        reverse_mapping[material.lower().strip()] = category

# Function to map each material to a category
def map_material(material):
    material = material.lower().strip()  # Convert to lowercase and strip extra spaces
    return reverse_mapping.get(material, 'Other material')  # Return category or 'Other' if not found

# Assuming 'df' is your DataFrame and the 'Material' column has the material names as single values per row
df['Material'] = df['Material'].apply(map_material)  # Apply mapping to 'Material' column

# Identify low-frequency categories (appear less than a threshold)
low_freq_threshold = 1
material_counts = df['Material'].value_counts()

# Map low-frequency categories to 'Other'
df['Material'] = df['Material'].apply(lambda x: 'Other' if material_counts[x] < low_freq_threshold else x)

# Print material category counts
print(df['Material'].value_counts())
```

```python
# Updated Hemline mapping dictionary
hemline_mapping = {
    'straight_hem': ['straight_hem'],
    'ribbed_hem': ['ribbed_hem','elasticated_hem'],
    'other_hem':['curved_hem','ruffle_hem','asymmetrical_hem']
}

# Function to map hemline values
def map_hemline(hemline_value):
    if isinstance(hemline_value, str):  # Check if the value is a string
        hemline_value = hemline_value.lower()  # Convert to lowercase for matching
        for key, values in hemline_mapping.items():
            if hemline_value in values:
                return key
    return 'other_hemline'  # Return 'other' if no match is found or if the value is not a string

# Apply the mapping
df['Hemline'] = df['Hemline'].map(map_hemline)

# Replace any NaN values (mismatched values) with 'other'
df['Hemline'] = df['Hemline'].fillna('other_hemline')

# Check the updated unique values in the 'Hemline' column
print(df['Hemline'].unique())
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Function to calculate Cramér's V
def cramers_v(chi2, n, r, c):
    return np.sqrt(chi2 / (n * min(r-1, c-1)))

# List of nominal variables
nominal_vars = ['Season', 'Collar', 'Neckline', 'Hemline', 'Sleeve Style',
                'Pattern', 'Material', 'Product Colour', 'Breathable', 'Lightweight', 'Water_Repellent']

# Initialize an empty dictionary to store the Cramér's V values
cramers_v_matrix = {}

# Loop through pairs of nominal variables and calculate Cramér's V
for var1 in nominal_vars:
    cramers_v_matrix[var1] = []
    for var2 in nominal_vars:
        if var1 != var2:
            # Create a contingency table
            contingency_table = pd.crosstab(df[var1], df[var2])
            
            # Perform Chi-square test
            chi2, p, dof, expected = chi2_contingency(contingency_table)
            
            # Calculate Cramér's V
            n = contingency_table.sum().sum()
            r, c = contingency_table.shape
            cramer_v_value = cramers_v(chi2, n, r, c)
            
            # Append the Cramér's V value to the list
            cramers_v_matrix[var1].append(cramer_v_value)
        else:
            # If the variable is the same, Cramér's V is 1
            cramers_v_matrix[var1].append(1)

# Convert the dictionary to a DataFrame for better visualization
cramers_v_df = pd.DataFrame(cramers_v_matrix, index=nominal_vars)

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(cramers_v_df, annot=True, cmap='coolwarm', vmin=0, vmax=1, fmt='.2f', linewidths=0.5)
plt.title("Association between nominal variables")
plt.show()
```
